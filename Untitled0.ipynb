```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caminho do arquivo CSV
file_path = '/content/relatorio_final.csv'  # Substitua pelo caminho correto

# Leitura do arquivo CSV
df = pd.read_csv(file_path)

# Configura√ß√£o para visualiza√ß√£o mais limpa
plt.style.use('seaborn-v0_8-darkgrid')

# 1. Quantidade de vendas por categoria de servi√ßo
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='categoria_servico', order=df['categoria_servico'].value_counts().index, palette='viridis')
plt.title('Quantidade de Vendas por Categoria de Servi√ßo')
plt.xlabel('Categoria de Servi√ßo')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Receita total por servi√ßo
df['receita_total'] = df['quantidade'] * df['preco_unitario']
receita_por_servico = df.groupby('servico')['receita_total'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
receita_por_servico.plot(kind='bar', color='skyblue')
plt.title('Receita Total por Servi√ßo')
plt.xlabel('Servi√ßo')
plt.ylabel('Receita Total (R$)')
plt.tight_layout()
plt.show()

# 3. Distribui√ß√£o das vendas por regi√£o
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='regiao', palette='pastel', order=df['regiao'].value_counts(85).index)
plt.title('Distribui√ß√£o das Vendas por Regi√£o')
plt.xlabel('Regi√£o')
plt.ylabel('Quantidade de Vendas')
plt.tight_layout()
plt.show()

# 4. Rela√ß√£o entre custo e receita total
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='custo', y='receita_total', hue='regiao', palette='Set2')
plt.title('Rela√ß√£o entre Custo e Receita Total')
plt.xlabel('Custo (R$)')
plt.ylabel('Receita Total (R$)')
plt.tight_layout()
plt.show()

# 5. Receita total por estado
receita_por_estado = df.groupby('estado')['receita_total'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
receita_por_estado.plot(kind='bar', color='orange')
plt.title('Receita Total por Estado')
plt.xlabel('Estado')
plt.ylabel('Receita Total (R$)')
plt.tight_layout()
plt.show()
```
